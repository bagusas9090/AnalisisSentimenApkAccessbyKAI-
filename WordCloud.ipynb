**Baris 2** Import wordcloud, pandas, dan matplotlib

```python
#import
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

**Baris 3** Import file csv yang berisi data teks tweet yang telah diberikan label sentimen

```python
#Membaca data csv dan menampilkan dataframe
df= pd.read_csv('/content/data_Stemming4.csv')
df
```

**Baris 4** Menyeleksi fitur atau kolom 'Stemming_text' dan 'Label'.

```python
#Menyeleksi fitur, atribut atau kolom yang akan digunakan

df= df[['Stemming_text','Label']]
df
```

**Baris 5** Merubah label string menjadi label numerik

```python
#Merubah label String menjadi Numerik menggunaan LabelEncoder
from sklearn.preprocessing import LabelEncoder

X = df['Stemming_text']
Y = df['Label']

target = {'Positif':0, 'Negatif':1, 'Netral':2}
df["Label"] = df["Label"].replace(target)

# Menampilkan DataFrame
print(df)
```

## **Langkah 2** Sentimen Negatif

**Baris 1** Memfilter label negatif

```python
#Visualisasi data dengan wordcloud

#Menampilkan data yang memiliki label negatif
senti_negatif = df[df['Label']==1]
senti_negatif
```

**Baris 2** Visualisasi wordcloud

```python
#Perintah menampilkan wordcloud
all_text_s1 = ' '.join(word for word in senti_negatif['Stemming_text'])
wordcloud = WordCloud(colormap='rainbow', mode='RGBA',background_color='white').generate(all_text_s1)

# Tampilkan word cloud
plt.figure(figsize=(55, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Sentimen Negatif")
plt.show()
```
